## ***Importing Libraries***









```python
import matplotlib.pyplot as plt
import cv2
from imutils.object_detection import non_max_suppression
from imutils import resize
import numpy as np
```

# **Creating instance of Histogram Of Oriented Gradients**

```python
hog=cv2.HOGDescriptor()
```

# ***Set SVM***

```python
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

## ***Setting the Image***

```python

img=cv2.imread('f.jpg')
img = resize(img, height=500)
rects, weights=hog.detectMultiScale(img, winStride=(4, 4), padding=(8, 8), scale=1.05)
copy=img.copy()
```

```python
for x, y, w, h in rects:
  cv2.rectangle(copy, (x, y), (x+w, y+h), (0,255,0), 2)
```

# ***Before Suppression***

```python
copy_rgb = cv2.cvtColor(copy, cv2.COLOR_BGR2RGB)
plt.imshow(copy_rgb)
plt.title('before suppression')
plt.axis('off')
plt.show()
cv2.waitKey(0)
```

```python
x=np.array([[x, y, x+w, y+h] for x, y, w, h in rects])
pick=non_max_suppression(x, probs=None, overlapThresh=0.65)
```

# ***After Suppression***

```python
for xa, ya, xb, yb in pick:
  cv2.rectangle(img, (xa, ya), (xb, yb), (0,0,255), 2)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title('after suppression')
plt.axis('off')
plt.show()
cv2.waitKey(0)
```
